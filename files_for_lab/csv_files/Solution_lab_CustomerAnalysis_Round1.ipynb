```python
# 5. Correct the values in the column 'customer lifetime value'. They are given as a percent, so divide them by 100 and change dtype to numerical type. (only 2 of the files need this step.)
file1.info()
file2.info()
file3.info()
```

```python
file1['customer_lifetime_value'] = pd.to_numeric(file1['customer_lifetime_value'], errors='coerce') / 100
display(file1.head())
display(file1.tail(7))
```

```python
# 6. Concatenate the three dataframes
file_gral = pd.concat([file1,file2,file3], axis=0)
display(file_gral.head())
display(file_gral.tail(7))
print(file_gral.shape)
```

```python
# 8. Which columns are categorical?
# In files 1 and 2, are categorical 'customer', 'state', 'gender', 'education', 'number_open_complaints', 'policy_type' and 'vehicle_class'. For file3, the same except 'number_open_complaints'.
file_gral.select_dtypes(['object'])
```

```python
# 9. Perform the data cleaning operations mentioned so far in class.
# 9.1 Delete the column education and the number of open complaints from the dataframe.
file_gral = file_gral.drop('education', axis=1)
display(file_gral.head())
display(file_gral.tail(7))
print(file_gral.shape)
```

```python
file_gral = file_gral.drop('number_open_complaints', axis=1)
display(file_gral.head())
display(file_gral.tail(7))
print(file_gral.shape)
```

```python
# 9.2 Check for duplicate rows in the data and remove if any.
file_gral.duplicated()
```

```python
#There are not duplicate rows. So, if I apply the drop function, the results are the same before and after of drop function.
print(file_gral.shape)
file_gral2 = file_gral.copy()
file_gral2.drop_duplicates()
print(file_gral2.shape)
```

```python
# 9.3 Filter out the data for customers who have an income of 0 or less. (Only show customers who HAVE an income.)

filtered = file_gral[file_gral['income']<=0]
display(filtered.head())
display(filtered.tail(7))
print(filtered.shape)
```
